### Visualising the data to aid us understand the data structure better. 

#### Importing the necessary libraries

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import seaborn as sns
import plotly.graph_objs as go
```

#### Accessing the data

```python
columns_to_use = ['userid', 'degree_1', 'degree_1_specializations', 'known_languages', 'key_skills', 'career_objective']
user_profile_ds = pd.read_csv('cleanedDatasets/cleaned_user_profile.csv',encoding='unicode_escape', usecols=columns_to_use)
user_ratings_ds = pd.read_csv('cleanedDatasets/cleaned_user_ratings.csv',encoding='unicode_escape')
course_ds = pd.read_csv('cleanedDatasets/cleaned_course_dataset.csv',encoding='unicode_escape')
```

Word Cloud for Known Languages: A word cloud can help visualize the most common languages known by users.

```python
# Convert all non-null elements to strings
text = ' '.join(user_profile_ds['known_languages'].dropna().astype(str))

# Generate the word cloud
wordcloud = WordCloud().generate(text)

# Display the word cloud
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

#### User Rating Dataset Visualizations

Heatmap of User Ratings: A heatmap can help visualize the rating distribution across different courses for all users.

```python
user_ratings_ds.head()
```

```python
#visualize ratings
rating_types = user_ratings_ds['rating'].value_counts()
x,y   = rating_types.index, rating_types.values
data  = go.Bar(x=x,y=y, text=y, textposition="outside")
Layout=go.Layout(title="Ratings Distributions",xaxis={'title':'Rating'},yaxis={'title':'Count'})
go.Figure(data,Layout)
```

#### Course Dataset Visualisations

```python
#visualize degree
rating_types = course_ds['degree_1'].value_counts()
x,y   = rating_types.index, rating_types.values
data  = go.Bar(x=x,y=y, text=y, textposition="outside")
Layout=go.Layout(title="Course Datset: Degree Distributions",xaxis={'title':'Degree'},yaxis={'title':'Count'})
go.Figure(data,Layout)
```

```python
#visualize degree specs
rating_types = course_ds['degree_1_specializations'].value_counts()
x,y   = rating_types.index, rating_types.values
data  = go.Bar(x=x,y=y, text=y, textposition="outside")
Layout=go.Layout(title="Course Dataset: Degree Specializations Distributions",xaxis={'title':'Degree Specializations'},yaxis={'title':'Count'})
go.Figure(data,Layout)
```
